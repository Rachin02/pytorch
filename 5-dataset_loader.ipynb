```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()                # scaling
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()                     # encoding
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train
```

```python
y_train
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
```

```python
train_dataset = CustomDataset(X_train_tensor, y_train_tensor)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor)
```

```python
train_dataset[10]
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size = 32, shuffle=True)
```

## Define Model

```python
import torch.nn as nn

class MyModel(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        out = self.linear(features)
        out = self.sigmoid(out)

        return out
```

```python
epochs = 25
learning_rate = 0.1
```

```python
# create model 
model = MyModel(X_train_tensor.shape[1])

# define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

# define loss
loss_function = nn.BCELoss()
```

## Training pipeliine

```python
# define loop
for epoch in range(epochs):
    for batch_feature, batch_label in train_loader:
        
        y_pred = model(batch_feature)    # Forward pass

        loss = loss_function(y_pred, batch_label.view(-1,1))  # Loss calculation

        optimizer.zero_grad()   # clear gradient

        loss.backward()    # Back propagation

        optimizer.step()   # parameters update

    
    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')   # print loss in each epoch
```

## Evaluation

```python
model.eval()    # set the model to evaluation mode
accuracy_list = []


with torch.no_grad():
    for batch_feature, batch_label in test_loader:

        y_pred = model(batch_feature)     # forward pass
        y_pred = (y_pred > 0.8).float()   # Convert probabilities to binary predictions

        # calculate accuracy for the current batch
        batch_accuracy = (y_pred == batch_label).float().mean().item()
        accuracy_list.append(batch_accuracy)

# calculate overall accuracy
accuracy = sum(accuracy_list)/ len(accuracy_list)
print(f'Accuracy: {accuracy:.4f}')



```
