## Matrix operations

```python
x = torch.randint(size = (2,3), low = 0, high = 10)
y = torch.randint(size = (3,2), low = 0, high = 10)
print(x)
print(y)
```

```python
# matrix multiplication

torch.matmul(x,y)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])
```

```python
# dot product
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(x, 0, 1)
```

## comparison operation

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than equal to

# less than equal to
```

## inplace operation

```python
m = torch.rand(2,3)
n = torch.rand(2,3)

print(m)
print(n)
```

```python
m.add_(n) ## n is added to m.no new memory is create
```

```python
m
```

```python
n
```

```python
torch.relu(m) #this will create different memory space
```

```python
m.relu_() # this will happen in the same memory space as m
```
